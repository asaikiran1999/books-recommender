#**BOOKS RECOMMENDATION SYSTEM**

**TYPES OF RECOMENDATION SYSTEMS**

**1.Popularity based recommendation system :** In this type of recommendation system we recommend items based its popularity under a specific formula running behind.

**2.Content based recommendation system :** In  this type of recommendation system we recommend items based on common attributes of content.

**3.Collaborative filtering based system :** In this type of recommendation system we recommend items based on common distribution under a batch of users. 

**4.Hybrid Recommendation System :** This is mixture of two or more rcommendation systems working at a time.

#**POPULARITY based recommendation system**

**1. Merging the books and rating dataframes**

```python
ratings_and_books = ratings.merge(books,on='ISBN')
ratings_and_books.head()
```

**2.Groupby book-title and count the number of book rattings** 

```python
x=ratings_and_books.groupby("Book-Title").count()['Book-Rating']
x.head()
```

**3.creating a seperate dataframe with books title and books rating count**

```python
x=x.reset_index()
x.rename(columns = {'Book-Rating':'ratings-Count'},inplace=True)
x.head()
```

```python
Ratings_count = x
Ratings_count.head()
```

**5.following the same above steps we createw a dataframe which shows average rating of particulatar book**

```python
x=ratings_and_books.groupby("Book-Title").mean()['Book-Rating'].reset_index()
x.rename(columns = {'Book-Rating':'average rating'},inplace=True)
Ratings_average=x
Ratings_average.head()
```

**6.merging the ratings_count and ratings _average dataframes into single dataframe**

```python
popular_recommend = Ratings_count.merge(Ratings_average ,on="Book-Title")
popular_recommend.head()
```

**7.filtering and sorting the dataset**

```python
popular_recommend = popular_recommend[popular_recommend['ratings-Count']>250].sort_values('average rating',ascending=False).head(50).reset_index()
popular_recommend.head()
```

**8.merging the popular_recommend with Books dataframe**

```python
x = popular_recommend.merge(books , on="Book-Title")
x.head()
```

**9.drop duplicates columns in merged dataframe**

```python
x = x.drop_duplicates('Book-Title')
x.shape
```

**10. Reindexing and showing the showing final top50 books**

```python
x=x[["Book-Title","ratings-Count","average rating","Book-Author","Image-URL-M"]].reset_index()
top50 = x
top50
```

#**COLLABORATIVE FILTERING based recommendation system**

**1.to groupby the top book ratting users**

```python
x=ratings_and_books.groupby('User-ID').count()['Book-Rating']>200
best_users=x[x].index
best_users.shape
```

**2.creating a new dataframe for top rating book users**

```python
filtered_ratings =ratings_and_books[ratings_and_books['User-ID'].isin(best_users)]
```

**3.to groupby the top famous books**

```python
y=filtered_ratings.groupby('Book-Title').count()['Book-Rating']>=50
famous_books=y[y].index
famous_books
```

**4.top famous books filter**

```python
filtered_famous = filtered_ratings[filtered_ratings['Book-Title'].isin(famous_books)]
filtered_famous
```

**5.constructing a pivot for the filtered data**

```python
pivot = filtered_famous.pivot_table(index ='Book-Title',columns='User-ID',values='Book-Rating')
pivot.fillna(0,inplace=True)
pivot
```

**6.calculating the cosine_similarity**

```python
from sklearn.metrics.pairwise import cosine_similarity
cosine_similarity(pivot)
```

```python
similarity_scores = cosine_similarity(pivot)
```

**7.building a book recommendation system**

```python
import numpy as np
def recommend(book_name):
  index = np.where(pivot.index == book_name)[0][0]
  distances = similarity_scores[index]
  similar_items =sorted(list(enumerate(similarity_scores[index])),key=lambda x:x[1],reverse=True)[1:6]
  for i in similar_items:
    print(pivot.index[i[0]]) 
```
